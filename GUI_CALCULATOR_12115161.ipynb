# GUI_CALCULATOR

```python
import tkinter as tk  #TKINTER MODULE IS USED FOR USE OF GRAPHICS

def calculate():
          # here we are getting the value ofnumbers
    num1 = int(entry_1.get()) 
    num2 = int(entry_2.get())
          #in operation we we define  which operation we want to perform on numbers
    operation = var.get()
    if operation == 1:
        result = num1 + num2
    elif operation == 2:
        result = num1 - num2
    elif operation == 3:
        result = num1 * num2
    else:
        result = num1 / num2
    result_label.config(text="Result: " + str(result))
    
# here we are created a objet of tkinter module
root = tk.Tk()
root.geometry("300x300") # here we defining the frame of gui interface
root.title("Calculator") # here we are defineing the title 
     

var = tk.IntVar()

lab=tk.Label(root,text="CALCULATOR",font=("Time New Roman",15,"bold"))
lab.place(x=10,y=0,height=55,width=300)

lab=tk.Label(root,text="Num1:: ",font=("Time New Roman",10,"bold"))
lab.place(x=12,y=59,height=15,width=50)

lab=tk.Label(root,text="Num2:: ",font=("Time New Roman",10,"bold"))
lab.place(x=12,y=85,height=15,width=50)

#The canvas widget is used to add the structured graphics to the python application
canvas = tk.Canvas(root, width=300, height=55) 
canvas.text = canvas.create_text(150, 30, text="......Caculator.....",font=("Time New Roman",25,"bold"))
canvas.pack()

entry_1 = tk.Entry(root,width=15, borderwidth=2) # here we are defeining the entry to take the value from user
entry_2 = tk.Entry(root,width=15, borderwidth=2)

#here we are using radio button for user to choose operation from divide ,multiply and divide
add_button = tk.Radiobutton(root, text="Add  ", variable=var, value=1,) 
subtract_button = tk.Radiobutton(root, text="Subtract", variable=var, value=2)
multiply_button = tk.Radiobutton(root, text="Multiply", variable=var, value=3)
divide_button = tk.Radiobutton(root, text="Divide", variable=var, value=4)

# here we defined a calculator button in which we are calling calculate function which we have define upper

calculate_button = tk.Button(root, text="Calculate", command=calculate)

result_label = tk.Label(root)

# here we are packing all the earlier call function
entry_1.pack()
entry_2.pack()

add_button.pack()
subtract_button.pack()
multiply_button.pack()
divide_button.pack()

calculate_button.pack()

result_label.pack()

root.mainloop() # here the tkinter  GUI interface is closed
```
